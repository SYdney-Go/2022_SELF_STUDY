## 1. 넘파이를 소개합니다

리스트 복습하기

```python
my_list = [10, "hello world", 20]
print(my_list[1])
```

```python
my_list2 = [[10, 20, 30], [40, 50, 60]]
print(my_list2)
print(my_list2[1][1])
```

넘파이 준비하기

```python
import numpy as np
print(np.__version__)
```

넘파이로 배열 만들기

```python
# array 함수로 2차원 배열 만들기
my_arr = np.array([[10, 20, 30], [40, 50, 60]])
print(my_arr)

# type 함수로 넘파이 배열인지 확인하기
print(type(my_arr))

# 넘파이 배열에서 요소 선택하기
print(my_arr[0][2])

# 넘파이 내장 함수 사용하기
print(np.sum(my_arr))
```

맷플롯립으로 그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
# 선그래프 그리기
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) #x축, y축 값 전달
plt.show()
```

```python
# 산점도 그리기
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

```python
#넘파이 배열로 산점도 그리기
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```
