##### 1. 재귀 호출 사용하기
재귀 호출 : 함수 안에서 자기 자신을 호출

```python
def hello():
    print("Hello, world!")
    hello() 

hello() # 이거 실행하면 끝도 없다. 주의... (-> 다행히 파이썬에서는 최대 재귀 깊이를 1000으로 정해놓았다. 이걸 초과하면 Recursion error 발생)
```

재귀호출에 종료 사건 만들기

```python
def hello(count):
    if count == 0:
        return

    print("Hello, world!", count)

    count -= 1
    hello(count)

hello(5)
```

##### 2. 재귀 호출로 팩토리얼 구하기

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```
